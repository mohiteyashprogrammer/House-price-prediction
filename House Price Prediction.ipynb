## scikit-learn Design

primarily, three type of object
1. Estimators- it estimates some perameter base on dataset. eg.imputer
it has a fit method and transform method.
Fit method - fits the dataset and calculates internal parameters
2. Transformers - transform method take inpute and returns output base on the 
learning from fit(). It also has a convence function called fit_transform()
which fits anf then transform.
3. Predictors - LinearRegression model is an example of predictor. fit() and
predict() are two common function. It also give function which will evaluate the prediction.

## Feature scaling

primarily, two type of feature scaling methods:
1. min- max scaling(normalization)
(value-min/(max-min)
sklearn provide a class called MinMaxScaler for this
2. Standardization
(value - mean)/ std
sklearn provide a class called standerdScaler for this

## Creating a pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    #..... add as many as you want in your pipeline
    ('std_scaler', StandardScaler()),
])
```

```python
housing_num = my_pipeline.fit_transform(housing)
```

```python
housing_num.shape
```

## Selecting a desire model for House Prediction

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
#model = LinearRegression()
#model = DecisionTreeRegressor()
model = RandomForestRegressor()
model.fit(housing_num, housing_labels)

```

```python
some_data = housing.iloc[:5]
```

```python
some_labels = housing_labels.iloc[:5]
```

```python
prepared_data = my_pipeline.transform(some_data)
```

```python
model.predict(prepared_data)
```

```python
list(some_labels)
```

## Evaluating The Model

```python
import numpy as np
from sklearn.metrics import mean_squared_error
housing_predictions = model.predict(housing_num)
mse = mean_squared_error(housing_labels, housing_predictions)
mse = np.sqrt(mse)
```

```python
mse
```

## Using better evalution technique - Cross Validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, housing_num, housing_labels, scoring= "neg_mean_squared_error", cv=10)
rmse_Scores = np.sqrt(-scores)
                
```

```python
rmse_Scores
```

```python
def print_scores(scores):
    print("scores:", scores)
    print("mean: ",scores.mean())
    print("standard deviation: ", scores.std())
```

```python
print_scores(rmse_Scores)
```

## Saving the model

```python
from joblib import dump, load
dump(model, 'house.joblib')
```

## Test the model on test data

```python
x_test = strat_test_set.drop("MEDV", axis = 1)
y_test = strat_test_set['MEDV'].copy()
x_test_prepared = my_pipeline.transform(x_test)
final_predictions = model.predict(x_test_prepared)
final_mse = mean_squared_error(y_test, final_predictions)
final_rmse = np.sqrt(final_mse)
#print(final_predictions, list(y_test))
```
